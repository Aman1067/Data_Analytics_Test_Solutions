```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
url = "https://drive.google.com/uc?id=1F_LOzrBtu4EAbmb2bx9SufprVDXk0Atm"
df = pd.read_csv(url)
```

```python
df.head()
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.set_index('date', inplace=True)
```

```python
rolling_avg = df['closing_price'].rolling(window=30).mean()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['closing_price'], label='Closing Price', color='blue', alpha=0.7)
plt.plot(rolling_avg.index, rolling_avg, label='30-Day Rolling Avg', color='red')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Stock Prices Over Time with 30-Day Rolling Average')
plt.legend()
plt.grid(True)
plt.show()
```
